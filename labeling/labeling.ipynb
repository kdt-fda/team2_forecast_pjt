```python
import ast # 마법 도구를 가져옵니다.

# 1. 'tokens' 칸에 있는 글자들을 진짜 리스트로 변신!
# 혹시라도 비어있거나(NaN) 이상한 값이 섞여 있을 수 있으니 안전하게 처리합니다.
def convert_to_list(x):
    if isinstance(x, str): # 만약 글자(str)라면
        return ast.literal_eval(x) # 진짜 리스트로 바꿔줘!
    return x # 이미 리스트라면 그냥 둬!

df['tokens'] = df['tokens'].apply(convert_to_list)

# 2. 잘 바뀌었는지 확인 (이제 <class 'list'>가 나와야 해요)
print(f"변신 후 타입: {type(df['tokens'].iloc[0])}")
```

### NBC 모델 학습

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import pandas as pd
import numpy as np

df_test = df_study.sample(n=min(500, len(df_study)), random_state=42)

# 2. 30번이 아니라 딱 3번만 반복해볼게요.
n_test_rounds = 3
bagging_results = []

for i in range(n_test_rounds):
    train_data = df_test.sample(frac=0.9, random_state=i)
    cv = CountVectorizer(tokenizer=lambda x: x, lowercase=False, preprocessor=lambda x: x)
    X_train = cv.fit_transform(train_data['tokens'])
    y_train = train_data['label']

    nbc = MultinomialNB()
    nbc.fit(X_train, y_train)

    dovish_idx = np.where(nbc.classes_ == 'dovish')[0][0]
    hawkish_idx = np.where(nbc.classes_ == 'hawkish')[0][0]

    prob_hawkish = np.exp(nbc.feature_log_prob_[hawkish_idx])
    prob_dovish = np.exp(nbc.feature_log_prob_[dovish_idx])

    polarity_ratio = prob_hawkish / (prob_dovish + 1e-10)
    words = cv.get_feature_names_out()
    score_df = pd.DataFrame({'word': words, f'score_{i}': polarity_ratio}).set_index('word')
    bagging_results.append(score_df)
    
    if (i+1) % 5 == 0:
        print(f"{i+1}번째 학습 완료")

final_lexicon = pd.concat(bagging_results, axis=1)
final_lexicon['polarity_score'] = final_lexicon.mean(axis=1)

print("\n--- [축하합니다!] 모든 단어의 평균 극성 점수가 산출되었습니다. ---")
```

```python
# intensity threshold = 1.3 기준으로 단어 사전 구축
hawkish_lexicon = final_lexicon[final_lexicon['polarity_score'] > 1.3]
dovish_lexicon = final_lexicon[final_lexicon['polarity_score'] < (1/1.3)]

print(f"hawkish 단어: {len(hawkish_lexicon)}개 발견")
print(f"dovish 단어: {len(dovish_lexicon)}개 발견")

hawkish_set = set(hawkish_lexicon.index)
dovish_set = set(dovish_lexicon.index)
```

```python
# 문장
def calculate_tone(n_hawkish, n_dovish):
    denominator = n_hawkish + n_dovish
    if denominator == 0:
        return 0  # 매/비 단어가 하나도 없으면 중립(0)
    return (n_hawkish - n_dovish) / denominator

# 문장 tone 지수 계산
df_study['n_h_feat'] = df_study['tokens'].apply(lambda x: len([w for w in x if w in hawkish_set]))
df_study['n_d_feat'] = df_study['tokens'].apply(lambda x: len([w for w in x if w in dovish_set]))
df_study['tone_s'] = df_study.apply(lambda row: calculate_tone(row['n_h_feat'], row['n_d_feat']), axis=1)

# 문장별 dovish, hawkish 여부 표시
df_study['is_h_sent'] = df_study['tone_s'] > 0
df_study['is_d_sent'] = df_study['tone_s'] < 0

```

```python
# 문서
# 문서별로 매파 문장 수와 비둘기파 문장 수 sum
doc_level = df_study.groupby(['doc_id', 'date']).agg(
    n_h_sents=('is_h_sent', 'sum'),
    n_d_sents=('is_d_sent', 'sum')
).reset_index()

doc_level['tone_i'] = doc_level.apply(lambda row: calculate_tone(row['n_h_sents'], row['n_d_sents']), axis=1)

```

```python
# 문장 톤 지수 비중 확인
tone_ratio = df_study['tone_s'].value_counts(normalize=True)
tone_percentage = tone_ratio * 100
print(tone_percentage)
```

```python
# 날짜별 지수 계산
daily_tone = doc_level.groupby('date')['tone_i'].mean().reset_index()
daily_tone.columns = ['date', 'date_tone']

display(daily_tone.head())
```
